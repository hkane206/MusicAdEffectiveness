```python
data = {
    
    "Average_Hue": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_average_hue_normalized.json",
    "Average_Intensities": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_average_intensities.json",
    "Average_Intensities_30": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_average_intensities_cropped_30percent.json",
    "Average_Intensities_60": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_average_intensities_cropped_60percent.json",
    "Average_Memorability": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_average_memorabilities.json",
    "Duration": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_Duration_new_raw.json",
    "Effectiveness": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_Effective_clean.json",
    "Entropy": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_entropy.json",
    "Exciting": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_Exciting_clean.json",
    "Funny": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_Funny_clean.json",
    "Language": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_Language_clean.json",
    "Median_Hue": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_median_hue_normalized.json",
    "Optical_Flow": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_optical_flow.json",
    "Sentiments": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_Sentiments_clean.json",
    "Shot_Boundary": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_shot_boundary.json",
    "Topics": "/Users/harrisonkane/Downloads/annotations_videos/video/cleaned_result/video_Topics_clean.json"

}
```

```python
json_total_data = {}

for key, value in data.items():
    
    with open(value) as f:
        json_data = json.load(f)
        json_total_data[key] = json_data

for feature, jsonfile in json_total_data.items():
        
        data_file = open("{}.csv".format(feature), "w")
        csv_writer = csv.writer(data_file)
        
        csv_writer.writerow(["Video", feature])
        
        for ad, value in jsonfile.items():
            
            csv_writer.writerow([ad, value])
            
data_file.close()
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ], axis = 1)
pd.set_option('display.max_columns', None)
combined_csv.drop(["NAME", "KEY", "CAMELOT", "BPM", "NAME", "NAME", "A", "B", "C", "D", "E", "F", "G", "Major", "Minor"], inplace = True, axis = 1)
combined_csv = combined_csv.iloc[: , :-1]
combined_csv.head()
#combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
test_csv = pd.read_csv("/Users/harrisonkane/Downloads/annotations_videos/BPM and Key.csv")
print("NAME" in test_csv)
test_csv.head()

final_csv = pd.concat([combined_csv, test_csv], axis = 1)
```
