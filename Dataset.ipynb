```python
json_total_data = {}

# Combine all json files downloaded from Hahn & Kovashka (2019)
# All json file paths stored in the dictionary data

for key, value in data.items():
    
    with open(value) as f:
        json_data = json.load(f)
        json_total_data[key] = json_data

#Write jsonfile into csv, creating columns alternating betwee video and feature
for feature, jsonfile in json_total_data.items():
        
        data_file = open("{}.csv".format(feature), "w")
        csv_writer = csv.writer(data_file)
        
        csv_writer.writerow(["Video", feature])
        
        for ad, value in jsonfile.items():
            
            csv_writer.writerow([ad, value])
            
data_file.close()
```

```python
test_csv = pd.read_csv("/Users/harrisonkane/Downloads/annotations_videos/BPM and Key.csv")
names = [i[:-4] for i in test_csv["NAME "]]
test_csv["NAME "] = names
test_csv.head()
keys = np.array([i[0:2] for i in test_csv["KEY"]])

# Sharp/Flat (either sharp (#) or neutral (" "))
sharp_flat = np.array([i[1] for i in keys])

# Last characters (either major or minor for all)
major_minor = [i[i.index(" ")+1:] for i in test_csv["KEY"]]

keys_modified = []

for key in keys:
    keys_modified.append(key[0])

# ONE-HOT ENCODING OF KEY
# First, letter name of key: key present = 1, key absent = 0
for i in range(len(keys_modified)):
    
    if keys_modified[i] == "A":
        test_csv.iloc[i, test_csv.columns.get_loc("A")] = 1
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("A")] = 0
        
    if keys_modified[i] == "B":
        test_csv.iloc[i, test_csv.columns.get_loc("B")] = 1
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("B")] = 0
        
    if keys_modified[i] == "C":
        test_csv.iloc[i, test_csv.columns.get_loc("C")] = 1
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("C")] = 0
        
    if keys_modified[i] == "D":
        test_csv.iloc[i, test_csv.columns.get_loc("D")] = 1
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("D")] = 0
        
    if keys_modified[i] == "E":
        test_csv.iloc[i, test_csv.columns.get_loc("E")] = 1
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("E")] = 0
        
    if keys_modified[i] == "F":
        test_csv.iloc[i, test_csv.columns.get_loc("F")] = 1
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("F")] = 0
    
    if keys_modified[i] == "G":
        test_csv.iloc[i, test_csv.columns.get_loc("G")] = 1
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("G")] = 0

# Second, major/minor: Major = 1, minor = 0
for i in range(len(major_minor)):
    
    if major_minor[i] == "major":
        test_csv.iloc[i, test_csv.columns.get_loc("Major/Minor")] = 1  
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("Major/Minor")] = 0

# Third, natural/sharp: Neutral = 0, sharp = 1 (all keys given in sharps)
for i in range(len(sharp_flat)):
    
    if sharp_flat[i] == " ":
        test_csv.iloc[i, test_csv.columns.get_loc("Sharp/Flat")] = 0
    else:
        test_csv.iloc[i, test_csv.columns.get_loc("Sharp/Flat")] = 1
```
